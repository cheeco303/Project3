```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("/Users/diboshbaruah/Desktop/Python/Hotel Reservations.csv")
print(data.head())
print()

data = data.drop(columns=['Booking_ID', 'booking_status'])

label_encoder = LabelEncoder()
for col in data.select_dtypes(include=['object']).columns:
    data[col] = label_encoder.fit_transform(data[col])

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

dbscan = DBSCAN(eps=0.5, min_samples=5)  # You may need to adjust eps and min_samples based on the data
clusters = dbscan.fit_predict(data_scaled)

data['Cluster'] = clusters

print("Cluster distribution:\n", data['Cluster'].value_counts())

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_scaled[:, 0], y=data_scaled[:, 1], hue=clusters, palette='viridis')
plt.title('DBSCAN Clustering of Hotel Reservations')
plt.show()

noise_points = (clusters == -1).sum()
print(f"Number of noise points: {noise_points}")
```

```python
# Build DBSCAN clustering algorithm on Hotel reservations data


import pandas as pd
from sklearn.preprocessing import StandardScaler , LabelEncoder
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns

# Load the hotel reservations data
hotel_data = pd.read_csv("/Users/diboshbaruah/Desktop/Python/Hotel Reservations.csv")


# To check for missing value 
missing_values = hotel_data.isnull().sum()
print(missing_values)

# Drop the 'Booking_ID' column as it's non-numeric and not needed for clustering
label_encoded_data = hotel_data.drop(columns=['Booking_ID', 'booking_status'])

# Use Label Encoding for categorical columns
label_encoders = {}
for col in ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']:
    le = LabelEncoder()
    label_encoded_data[col] = le.fit_transform(label_encoded_data[col])
    label_encoders[col] = le

# Normalize the data using StandardScaler
scaler = StandardScaler()
scaled_encoded_data = scaler.fit_transform(label_encoded_data)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters_encoded = dbscan.fit_predict(scaled_encoded_data)

# Add cluster labels to the original data
hotel_data['Cluster'] = clusters_encoded

# Display the cluster distribution
cluster_distribution_encoded = hotel_data['Cluster'].value_counts()
print(cluster_distribution_encoded)

# Save the results to a new CSV file
hotel_data.to_csv('Hotel_Reservations_with_Encoded_Clusters.csv', index=False)

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_scaled[:, 0], y=data_scaled[:, 1], hue=clusters, palette='viridis')
plt.title('DBSCAN Clustering of Hotel Reservations')
plt.show()
```
